## Introduction to NumPy

* NumPy was created in 2005 by Travis Oliphant. It is an open source project and you can use it for free.
* One of the reasons NumPy is so important in Python is because it is **designed for efficiency on large arrays of data**.
* NumPy internally stores data in a **contiguous block of memory**, independent of other built-in Python objects.
* NumPy arrays also use much **less memory than built-in Python sequences**.
* NumPy operations **perform complex computations on entire arrays without the need for Python for loops**, which can be slow for large sequences.
* NumPy is faster than regular Python code because its C-based algorithms avoid **overhead present with regular interpreted Python code**.
* NumPy’s library of algorithms written in the C language can operate on this memory **without any type checking or other overhead**.
* The array object in NumPy is called `ndarray`, it provides a lot of supporting functions that make working with `ndarray` very easy.
* Arrays are very frequently used in data science, **where speed and resources are very important**.
* NumPy-based algorithms are generally 10 to 100 times faster (or more) than their pure Python counterparts and use significantly less memory.

## Memory and speed aspect of Numpy compared to Lists


One of the main reasons that numpy is the foundation of data science in Python is that it does numerical computations incredibly efficiently both in the sense that it is fast and also in the sense that it is good at not using up all of your computer's memory.
To illustrate, let's create a list with all the numbers from one to one hundred million and sum them up. We will do this twice—once using regular Python tools, and once using numpy.

But that's not all -- regular Python required much more memory to store its list of numbers than was required by numpy:

asizeof.asizeof() gets the size of an object and all of its contents in bytes.

- asizeof.asizeof() (from the pympler library):- This function goes beyond the capabilities of sys.getsizeof() and .__sizeof__() by calculating the total memory footprint of an object, including all referenced objects.
- It is particularly useful for analyzing complex objects with nested structures, as it accounts for the memory usage of all components within the object.It is used for a detailed analysis of the memory usage of an object and **its references.**


```python
import numpy as np
try:
  from pympler import asizeof
except:
  !pip install pympler
  from pympler import asizeof
# Memory usage of NumPy array
array = np.array([1, 2, 3, 4, 5])
print('Outputs memory of array size in bytes:', asizeof.asizeof(array))
# Memory usage of Python list
lst = [1, 2, 3, 4, 5]
print('Outputs memory size of list in bytes:', asizeof.asizeof(lst))
```

```python
# Memory usage of NumPy array
array = np.array([1, 2, 3, 4, 5], dtype = 'uint8')
print('Outputs memory of array size in bytes:', asizeof.asizeof(array))
# Memory usage of Python list
lst = [1, 2, 3, 4, 5]
print('Outputs memory size of list in bytes:', asizeof.asizeof(lst))
```

* To measure how long each of these operations takes, we will use the time() function from the time module in the Python standard library.
* When run, time() returns the current time measured in the number of seconds that have passed since January 1st, 1970 UTC.
* So if we record the time before we start running our code, then record the time when the code finishes, then the difference is the amount of time (in seconds) that the code took to run.

```python
import numpy as np
import time

# Large NumPy array
large_array = np.arange(1_000_000_0)

start = time.time()
result = large_array * 2  # Vectorized operation
end = time.time()
numpy_total = end - start
print("NumPy Time:", numpy_total)

# Large Python list
large_list = list(range(1_000_000_0))

start = time.time()
result = [x * 2 for x in large_list]  # Iterative operation
end = time.time()
python_total = end - start
print("List Time:", python_total)

# Speed comparison
print(f"NumPy was {python_total / numpy_total:.1f}x faster!")
```

## Numpy Data Types

- The `dtype` object in NumPy holds metadata that allows the ndarray to interpret a block of memory as a specific data type.
- NumPy's data types are a key factor in its versatility, enabling seamless interaction with data from external systems. They often provide a direct mapping to underlying disk or memory formats, which facilitates binary data read/write operations and integration with code written in low-level programming languages like C or FORTRAN.
- Numerical data types in NumPy follow a consistent naming convention, consisting of a type name (e.g., `float` or `int`) followed by the number of bits allocated per element.
- A double-precision floating-point value (as represented in Python’s `float` object) requires 8 bytes or 64 bits of memory. In NumPy, this data type is referred to as `float64`.

###  Memory Optimization with dtypes & astype()

Optimizing memory usage in NumPy often revolves around understanding and effectively utilizing `dtypes` and the `astype()` method.

Here’s an overview of some common NumPy data types and their representations:
* Signed and unsigned 32-bit integers: `int32 (i4)`, `uint32 (u4)`
* Signed and unsigned 64-bit integers: `int64 (i8)`, `uint64 (u8)`

* Standard double-precision floating point (corresponds to * Python's float object): `float64 (f8 or d)`

### Type Promotion and Type Casting in NumPy

* Arrays are **designed to hold elements of a single data type**. However, if you provide a list containing **elements of varying data types to `np.array`, NumPy employs a process called `Type Promotion` to convert all elements into a uniform type within the array**.
* This conversion follows a hierarchical order, where **lower-ranking types can be promoted to higher-ranking ones**, but the reverse is not possible. The hierarchy is as follows:

$$Boolean → integer → float → string$$

* When Python encounters mixed data types, **it minimizes the promotion required to unify all elements under the same data type**, adhering to this hierarchy.

* Although certain special cases allow backward conversion—such as turning the character "5" into the integer 5 it is not universally feasible to convert characters to numeric types. Therefore, **Python consistently promotes types only in directions that are guaranteed to work**.



```python
np.array(["Nick", 42])
```

```python
# Combining Boolean and float:
np.array([1, 2.4, True])
```

```python
# Combining Boolean, float, and string:
np.array([True, 42, "Julio"])
```

```python
# Combining floats and integers:
my_vector = np.array([42, 47, 1.618])
my_vector, my_vector.dtype
```

```python
# Combining integers, Booleans, and floats:
my_new_vector = np.array([47, True, 3.14])
my_new_vector, my_new_vector.dtype
```
